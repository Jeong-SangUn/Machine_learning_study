```python
import pandas as pd
```

```python
GDP_df = pd.read_csv('./data/gdp2018.csv')
```

```python
GDP_df.head()
```

```python
GDP = GDP_df.sort_values('GDP',ascending=False)
```

```python
GDP['Ranking'] = [x for x in range(1,len(GDP)+1)]
```

```python
GDP
```

```python
GDP[GDP['Ranking']<=5][['Code','Ranking']]
```

```python
GDP[GDP['Nation'].isin(['China','Japan','Korea Rep.'])][['Nation','Ranking','GDP']]
```

```python
GDP[GDP['Ranking'].isin([x for x in range(1,len(GDP)+1,2)])][['Code','Ranking','GDP']]
```

```python
GDP_top10 = GDP[GDP['Ranking']<=10]
```

```python
GDP_top10[GDP_top10['Ranking'].isin([x for x in range(0,len(GDP_top10)+1,2)])][['Code','Ranking','GDP']]
```
