```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# 평균이 0이고 표준편차가 1인 정규분포 출력
a = np.random.randn(100000)
```

```python
print(a.mean())
print(a.std())
```

```python
np.random.rand(50)
```

```python
plt.hist(a, bins=25 ,color='k', alpha=0.1, label='histogram')
plt.legend(loc='best')
```

```python
fig, axes = plt.subplots(2,2, sharex = True, sharey= True, figsize = (12,8))

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(1000), bins=20, color='k', alpha=0.2)
        axes[i,j].set_xlabel('X{}{}'.format(i,j))
        axes[i,j].set_ylabel('value')
        
plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
