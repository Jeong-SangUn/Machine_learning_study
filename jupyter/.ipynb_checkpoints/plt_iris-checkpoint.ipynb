## 1. sklearn 라이브러리의 iris 데이터 불러오기

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
print('iris_dataset key:{}'.format(iris_dataset.keys()))
```

```python
print(iris_dataset['target'])
print(iris_dataset['target_names'])
```

```python
print(iris_dataset['DESCR'])
```

# 3. 다양한 머신러닝 분류 기법

## 3.1 k-nea

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# 1. 모든 feature 사용

X = iris_dataset.data
y = iris_dataset.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 35)
```

```python
np.bincount(y_train)
```

```python
# 모델 객체 생성
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# 모델 훈련
clf.fit(X_train, y_train)
```

```python
# 예측
predict_label = clf.predict(X_test)
```

```python
predict_label == y_test
```

```python
print('test accuracy {:2f}'.format(np.mean(predict_label == y_test)))
```

```python
plt.scatter(X_train[:,0], X_train[:,1], edgecolor = 'k', c = y_train)
```

```python
plt.scatter(X_train[:,2], X_train[:,3], edgecolor = 'k', c = y_train)
```

```python
print('트레인 세트 정확도: {:.2f}'.format(clf.score(X_train,y_train)))
```

```python
print('테스트 세트 정확도: {:.2f}'.format(clf.score(X_test,y_test)))
```

#### [연습] 위에서 시각화한 아아래 그래프의 레이블 레전드를 표시해 보시오

```python
plt.scatter(X_train[:,2][y_train==0], X_train[:,3][y_train==0], edgecolor='k',s=50 , label='setosa')
plt.scatter(X_train[:,2][y_train==1], X_train[:,3][y_train==1], edgecolor='k',s=50 , label='versicolor')
plt.scatter(X_train[:,2][y_train==2], X_train[:,3][y_train==2], edgecolor='k',s=50 , label='virginica')
plt.legend()
```
